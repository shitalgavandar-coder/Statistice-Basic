## 1. What is the difference between descriptive statistics and inferential statistics? Explain with examples. 


1 Descriptive Statistics

Definition:
Descriptive statistics summarize and describe the main features of a dataset.
It helps to organize, present, and analyze data in a meaningful way — but does not draw conclusions beyond the data available.

Purpose:
To describe what is happening in the data.

Examples:

Calculating the average marks of students in a class.

Finding the median income of people in a city.

Showing data using graphs, charts, tables, or percentages.

Example Explanation:
Suppose you surveyed 100 students and found that their average height is 165 cm.
 This is descriptive because it only describes your sample data.

    
2 Inferential Statistics

Definition:
Inferential statistics use a sample of data to make predictions, generalizations, or conclusions about a larger population.
It often involves probability theory.

Purpose:
To infer or make predictions about a population based on sample data.

Examples:

Predicting the average height of all students in a country using data from 100 students.

Conducting a hypothesis test to see if a new medicine is effective.

Calculating confidence intervals or performing regression analysis.

Example Explanation:
If you collect data from 100 students and then estimate the average height of all students in India,
→ you are using inferential statistics because you’re making a generalization beyond the sample.

## 2. What is sampling in statistics? Explain the differences between random and stratified sampling. 


In statistics, sampling is the process of selecting a small group (sample) from a large population to collect data and draw conclusions about the whole population.

Because studying every member of a population is often time-consuming, expensive, or impossible, we study a representative sample instead.

Example of Sampling:

Suppose you want to find the average height of students in a university with 10,000 students.
It’s not practical to measure all 10,000 students — so you select 200 students randomly and calculate their average height.
→This group of 200 students is your sample.

Types of Sampling:

There are many sampling methods, but two common ones are Random Sampling and Stratified Sampling.

 1. Random Sampling

Definition:
In random sampling, every individual in the population has an equal chance of being selected.

Purpose:
To eliminate bias and ensure that the sample represents the population fairly.

Example:
You have a list of 1,000 students.
You use a random number generator to select 100 students from the list.
→ Each student has an equal chance of selection.

Advantages:

Simple and unbiased

Easy to understand

Disadvantages:

May not represent all subgroups (e.g., gender, age, department)

 2. Stratified Sampling

Definition:
In stratified sampling, the population is divided into subgroups (strata) based on certain characteristics (like gender, age, income, etc.), and then samples are taken randomly from each stratum.

Purpose:
To ensure that every subgroup is represented proportionally in the sample.

Example:
Suppose a university has:

60% female students

40% male students

To ensure proper representation, you divide the population into two strata (male and female) and then randomly select samples from each group in the same proportion (e.g., 60 females and 40 males out of 100 students).

Advantages:

Ensures representation of all subgroups

More accurate and reliable results

Disadvantages:

More complex to organize

Requires prior knowledge of the population structure

 2. Median

Definition:
The median is the middle value of the data when arranged in ascending or descending order.
If there is an even number of values, the median is the average of the two middle values.

Example:
Data: 5, 10, 15, 20, 25
→ Middle value = 15, so median = 15

If data: 5, 10, 15, 20
→ Median = (10 + 15) / 2 = 12.5

Interpretation:
→ Half the data is below 15, and half is above 15.

 3. Mode

Definition:
The mode is the value that appears most frequently in the dataset.
A dataset can have one mode (unimodal), more than one mode (bimodal/multimodal), or no mode.

Example:
Data: 4, 5, 5, 6, 7, 7, 7, 8
→ Mode = 7 (appears most often)

Interpretation:
→ The most common value is 7.

    The measures of central tendency — mean, median, and mode — are important because they help us summarize and understand large sets of data using a single representative value.
Here’s why they are essential 

 1. They Summarize Data Simply

These measures condense a large amount of data into a single value, making it easy to understand the overall pattern or trend.

Example: Instead of listing the salaries of 500 employees, we can simply say the average salary is ₹45,000.

 2. They Help in Comparison

They allow us to compare different groups or time periods quickly.

Example: If the mean test score of Class A is 80 and Class B is 70, we can conclude that Class A performed better on average.

 3. They Show the “Center” or Typical Value

Each measure tells us something about the typical or central value of data:

Mean → the overall average

Median → the middle position (useful when there are outliers)

Mode → the most frequent or popular value

Example: In a clothing store, the mode size helps decide which size to stock more.

 4. They Are Useful in Decision-Making

Businesses, governments, and researchers use them to make informed decisions.

Example:

A company may use mean sales to plan targets.

A school may use median marks to evaluate student performance.

A retailer may use mode to decide which product size or color sells most.

 5. They Form the Basis for Further Statistical Analysis

These measures are the foundation for advanced statistics, such as variance, standard deviation, correlation, and regression.
You can’t analyze data distribution or variability without first knowing its center.

## 4. Explain skewness and kurtosis. What does a positive skew imply about the data? 

1. Skewness

Definition:

Skewness measures the asymmetry (lack of symmetry) of a data distribution around its mean.

If the data is perfectly symmetrical, skewness = 0.

If the data leans to one side, it is skewed.

2. Kurtosis

Definition:

Kurtosis measures the "peakedness" or flatness of a data distribution compared to a normal (bell-shaped) curve.
It tells us how heavy or light the tails of the distribution are.

A positive skew (right skew) means:

The tail of the distribution is longer on the right side.

Most values are small, but a few large values pull the mean to the right.

Mean > Median > Mode

 Example:
If most people earn ₹30,000–₹50,000 per month, but a few earn ₹5,00,000+,
→ The data is positively skewed (right-skewed).

## 5. Implement a Python program to compute the mean, median, and mode of a given list of numbers. numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28] 

```python
# Import statistics module
import statistics as stats

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate mean, median, and mode
mean_value = stats.mean(numbers)
median_value = stats.median(numbers)
mode_value = stats.mode(numbers)

# Display results
print("Numbers:", numbers)
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

 ## 6. Compute the covariance and correlation coefficient between the following #two datasets provided as lists in Python: 
## list_x = [10, 20, 30, 40, 50] 
## list_y = [15, 25, 35, 45, 60]


```python
import numpy as np


list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

x = np.array(list_x)
y = np.array(list_y)


cov_matrix = np.cov(x, y)  
covariance = cov_matrix[0, 1] 


correlation = np.corrcoef(x, y)[0, 1]


print("Covariance:", covariance)
print("Correlation Coefficient:", correlation
)
```

## 7. Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]


```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create a boxplot
plt.boxplot(data, vert=True, patch_artist=True, showmeans=True)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.grid(True)
plt.show()

# Identify outliers using IQR method
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Data:", data)
print("Q1:", Q1, "Q3:", Q3, "IQR:", IQR)
print("Lower Bound:", lower_bound, "Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

## 8. You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales. 
## Explain how you would use covariance and correlation to explore this relationship. 
## Write Python code to compute the correlation between the two lists: 
## advertising_spend = [200, 250, 300, 400, 500]   
## daily_sales = [2200, 2450, 2750, 3200, 4000]

1. Using Covariance and Correlation
Covariance

Covariance measures how two variables change together.

Positive covariance → As advertising spend increases, daily sales tend to increase.

Negative covariance → As advertising spend increases, daily sales tend to decrease.

Limitation: Covariance magnitude depends on the units of the variables, so it’s hard to interpret directly.

Correlation

Correlation coefficient (Pearson correlation) measures the strength and direction of the linear relationship between two variables.

Value range: −1 to +1

+1 → Perfect positive linear relationship

−1 → Perfect negative linear relationship

0 → No linear relationship

Correlation is unitless and easier to interpret than covariance.

In this scenario:

If correlation is close to +1, increasing advertising spend is strongly associated with higher daily sales.

If correlation is near 0, advertising spend has little to no linear effect on sales.

3. Interpretation

Covariance = 53750 → Positive value, indicating that as advertising spend increases, daily sales also increase.

Correlation ≈ 0.987 → Very strong positive linear relationship.

This means higher advertising spend is strongly associated with higher sales.

Actionable insight: Increasing advertising budget is likely to increase daily sales.

```python
import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to NumPy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Calculate covariance
cov_matrix = np.cov(x, y, bias=False)
covariance = cov_matrix[0, 1]

# Calculate correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Display results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

## Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product. 
## Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use. 
## Write Python code to create a histogram using Matplotlib for the survey data: survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

1. Summary Statistics and Visualizations

To understand the distribution of survey scores (1–10 scale), you can use:

Summary Statistics

Mean → Shows the average satisfaction score.

Median → Middle value; helps if data is skewed.

Mode → Most common score.

Standard Deviation (SD) → Measures the spread of scores; small SD → scores are close to the mean, large SD → scores vary widely.

Minimum & Maximum → Identify the range of responses.

Visualizations

Histogram → Shows frequency of each score; easy to see the overall distribution.

Boxplot → Identifies median, quartiles, and potential outliers.

Bar chart → Can also show counts of each score

```python
import matplotlib.pyplot as plt

# Survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Create histogram
plt.hist(survey_scores, bins=range(4, 12), edgecolor='black', color='skyblue', alpha=0.7)
plt.title("Customer Satisfaction Survey Scores")
plt.xlabel("Survey Score")
plt.ylabel("Frequency")
plt.xticks(range(4, 11))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Optional: Calculate basic statistics
import statistics as stats

mean_score = stats.mean(survey_scores)
median_score = stats.median(survey_scores)
mode_score = stats.mode(survey_scores)
std_dev = stats.stdev(survey_scores)

print("Mean:", mean_score)
print("Median:", median_score)
print("Mode:", mode_score)
print("Standard Deviation:", round(std_dev, 2))
```
